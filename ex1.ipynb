```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x, y1, label = "X axis")
plt.plot(x, y2, label = "Y axis")
plt.plot()

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("DISTRIBUTION PER SEMISTER")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x1 = ["BSE","BBIT", "BSCIT", "BCS", "BCM"]
y1 = [200, 400, 250, 180, 180]

plt.bar(x1, y1, label="yellow bar", color = 'y')
plt.plot()
plt.title("INTAKE AS PER THE COURSES")
plt.xlabel("COURSE")
plt.ylabel("NUMBER OF STUDENTS")
plt.legend()
plt.show()
```
